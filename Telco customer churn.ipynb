                                                  Feature Engineering & Data Split

```python
#Models learn from features (X), Models predict target (y)

x= df.drop("Churn", axis= 1)
y= df["Churn"]
```

```python
x.shape
```

```python
y.shape
```

```python
# Identify Categorical vs Numerical Columns

categorical_cols= x.select_dtypes(include="object").columns
numerical_cols= x.select_dtypes(exclude="object").columns

print("Categorical columns:", categorical_cols)
print("Numerical columns:", numerical_cols)
```

```python
# One-Hot Encode Categorical Variables: It converts each category into a separate numeric column.
# used one-hot encoding to convert categorical features into numeric form so the model can process them. I dropped the first category to avoid redundant information.

x= pd.get_dummies(x, drop_first=True)
```

```python
# more columns — this is expected and correct.

x.shape
```

```python
# Train / Test Split

from sklearn.model_selection import train_test_split
```

```python
test_size= 0.2        # 20% data for testing
random_state= 42      # 42 is just a popular random no.
stratify_column= y    # IMP:- Use target variable for balanced split (Keeps class balance in train & test sets.)
```

```python
#This code splits the dataset into training and testing sets while keeping class balance and ensuring reproducible results.
# Always use random_state and stratify for classification problems.

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= test_size, random_state= random_state, stratify= stratify_column)
```

```python
# Verify the Split
#check size

print(x_train.shape)
print(x_test.shape)
```

```python
# Check churn distribution

print(y_train.value_counts(normalize= True))
print(y_test.value_counts(normalize= True))

#The percentages should be almost identical.
#This confirms no sampling bias.
```

```python
# After removing duplicates, I separated features and target, encoded categorical variables using one-hot encoding, and performed a stratified train-test split to preserve class imbalance
```

                                                            Model building

LOGISTIC REGRESSION

```python
# start with Logistic Regression as a baseline before moving to more complex models.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
#from sklearn.pipeline import Pipeline

#LogisticRegression → the model
#StandardScaler → scales numeric features
#Pipeline → keeps preprocessing + model together (clean & safe)
```

```python
# scale the feature

scaler= StandardScaler()

x_train_scaled= scaler.fit_transform(x_train)    #fit_transform() → learn scaling from training data
x_test_scaled= scaler.transform(x_test)          #transform() → apply same scaling to test data  ---(avoids data leakage)
```

```python
# create logistic regression model

log_reg= LogisticRegression(max_iter= 1000)
```

```python
# Train the model

log_reg.fit(x_train_scaled, y_train)
```

```python
#make predictions

y_pred= log_reg.predict(x_test_scaled)
```

```python
#I scaled the numerical features using StandardScaler, trained a Logistic Regression model on the scaled data, and then made predictions on the test set.
```

```python
from sklearn.metrics import classification_report

#evaluate the model

print(classification_report(y_test, y_pred))
```

```python
#Model is very good at predicting non-churn
#Model misses ~47% of churners (recall = 0.53)
#Accuracy looks decent, but accuracy is misleading here

#This is NORMAL for Logistic Regression on churn data
#This is exactly why it’s called a baseline model

#explaination:- “The Logistic Regression model achieved about 80% accuracy, but recall for churn customers was lower. Since churn prediction is a recall-critical problem, this baseline helped identify the need for more complex models like Random Forest or XGBoost.”
```

RANDOM FOREST

```python
#After establishing a Logistic Regression baseline, I used Random Forest to capture non-linear patterns and improve churn recall.
```

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
rf_model= RandomForestClassifier(n_estimators= 100, random_state= 42)   #n_estimators=100 → 100 decision trees
```

```python
#train the model

rf_model.fit(x_train, y_train)
```

```python
#make predictions

y_pred_rf= rf_model.predict(x_test)
```

```python
#model evaluation

print(classification_report(y_test, y_pred_rf))
```

CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred_rf)
```

## Although Random Forest achieved similar accuracy, the confusion matrix showed a higher number of false negatives, meaning more churners were missed. Since churn prediction is recall-sensitive, Logistic Regression performed better as a baseline.

                                             Fix Random Forest for Churn

```python
# Use class weights: Tell the model- “Churn mistakes are more expensive.”
#class_weight="balanced" → penalize churn mistakes
#max_depth=10 → prevent overfitting
#min_samples_leaf=20 → smoother decisions

rf_model_balanced= RandomForestClassifier(n_estimators= 200, random_state= 42, class_weight= "balanced", max_depth= 10, min_samples_leaf= 20)
```

```python
# train

rf_model_balanced.fit(x_train, y_train)
```

```python
#predict

y_pred_rf_bal= rf_model_balanced.predict(x_test)
```

```python
# evaluate

print(classification_report(y_test, y_pred_rf_bal))
```

                                                Threshold Tuning Code

```python
# get churn probabilities

y_prob= rf_model_balanced.predict_proba(x_test)[:, 1]
```

```python
threshold= 0.3
y_pred_custom= (y_prob >= threshold).astype(int)
print(classification_report(y_test, y_pred_custom))
```

## I tuned the decision threshold to prioritize recall for churn customers, as missing churners is more costly than false positives.

## Lowering the classification threshold increased churn recall significantly while reducing accuracy. This trade-off is acceptable because missing churners is more costly than falsely flagging non-churn customers.

## The optimal threshold depends on business cost. I experimented with thresholds and selected one that maximized recall while keeping false positives manageable.
